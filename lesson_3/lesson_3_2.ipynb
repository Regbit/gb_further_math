```python
from math import sqrt, radians

class Vector2:
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __str__(self):
        return f"({self.x}; {self.y})"
    
    @classmethod
    def dist(cls, v1, v2):
        return round(sqrt((v2.x - v1.x)**2 + (v2.y - v1.y)**2), 8)

x_ortho = lambda v, rad, a, b: round((v.x-a)*np.cos(rad) + (v.y-b)*np.sin(rad), 8)
y_ortho = lambda v, rad, a, b: round(-(v.x-a)*np.sin(rad) + (v.y-b)*np.cos(rad), 8)

v1 = Vector2(3, 4)
v2 = Vector2(1, 5)

# Повернём на angle градусов и сместим на (a, b)
angle = radians(45)
a = 2
b = -1

v1_ = Vector2(x_ortho(v1, angle, a, b), y_ortho(v1, angle, a, b))
v2_ = Vector2(x_ortho(v2, angle, a, b), y_ortho(v2, angle, a, b))

print(f"{v1}, {v2}")
print(f"{v1_}, {v2_}")
print(f"{Vector2.dist(v1, v2)} == {Vector2.dist(v1_, v2_)}")
```

```python
class Polar2:
    
    def __init__(self, r, rad):
        self.r = r
        self.rad = rad
        
    def to_dec(self):
        return Vector2(self.r * np.cos(self.rad), self.r * np.sin(self.rad))
```

```python
print(Polar2(3, radians(45)).to_dec())
print(Polar2(7, radians(0)).to_dec())
print(Polar2(6, radians(30)).to_dec())
```

```python
x = np.linspace(-5, 5, 50)
plt.plot(x, x**2 - 1)
plt.plot(x, (np.exp(x) + x - 1) / x)
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.ylim(1, 8)
plt.xlim(-2, 3.5)
```

```python
from scipy.optimize import fsolve

def eq(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x * (1 - y) - 1)
    
x1, y1 = fsolve(eq, (-1, 2))
print(x1, y1)

x2, y2 = fsolve(eq, (2, 6))
print(x2, y2)
```

```python
x = np.linspace(-5, 5, 50)
plt.plot(x, x**2 - 1)
plt.plot(x, (np.exp(x) + x - 1) / x)

plt.plot(x1, y1, marker='o', markersize=5, color="red")
plt.plot(x2, y2, marker='o', markersize=5, color="red")

plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.ylim(1, 8)
plt.xlim(-2, 3.5)
```
