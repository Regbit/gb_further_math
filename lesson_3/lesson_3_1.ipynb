```python
from math import sqrt

class Vector3:
    
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
        
    def len(self):
        return sqrt(self.x**2 + self.y**2 + self.z**2)
    
```

```python
Vector3(1, 3, 5).len()
```

```python
Vector3(-9, 4, 0).len()
```

```python
Vector3(10, 10, 10).len()
```

```python
from math import sin, cos, radians

x = []
y = []
r = 5

for ang in range(-90, 90):
    x.append(r * sin(radians(ang)))
    y.append(r * cos(radians(ang)))

for ang in range(-90, 90):
    x.append(r * sin(radians(-ang)))
    y.append(-r * cos(radians(-ang)))

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.gca().set_aspect('equal', adjustable='box')
```

```python
x = []
y = []
a = 5
b = 2

for ang in range(-90, 90):
    x.append(a * sin(radians(ang)))
    y.append(b * cos(radians(ang)))

for ang in range(-90, 90):
    x.append(a * sin(radians(-ang)))
    y.append(-b * cos(radians(-ang)))

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.gca().set_aspect('equal', adjustable='box')
```

```python
x1 = []
y1 = []
x2 = []
y2 = []
a = 3
b = 2

step = 200
corr = 25

for i in range(step):
    x_i = (step-i)/corr
    if x_i >= a:
        x1.append(x_i)
        y1.append(sqrt(((x_i**2) / (a**2) - 1) * (b**2)))
        
for i in range(step):
    x_i = i/corr
    if x_i >= a:
        x1.append(x_i)
        y1.append(-sqrt(((x_i**2) / (a**2) - 1) * (b**2)))

for i in range(step):
    x_i = (step-i)/corr
    if x_i >= a:
        x2.append(-x_i)
        y2.append(sqrt((((-x_i)**2) / (a**2) - 1) * (b**2)))
        
for i in range(step):
    x_i = i/corr
    if x_i >= a:
        x2.append(-x_i)
        y2.append(-sqrt((((-x_i)**2) / (a**2) - 1) * (b**2)))
        

plt.plot(x1, y1)
plt.plot(x2, y2)
plt.xlabel("x")
plt.ylabel("y")
plt.gca().set_aspect('equal', adjustable='box')
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

fig = figure()
ax = Axes3D(fig)

x = np.arange(-5, 5, 2)
y = np.arange(-5, 5, 2)
x, y = np.meshgrid(x, y)
z1 = -2*x + 0.25*y
z2 = -2*x + 0.25*y + 20

ax.plot_surface(x, y, z1)
ax.plot_surface(x, y, z2)

show()
```
